# Model Traning

```python
X = read_csv.drop(['diagnosis'],axis = 1)
```

```python
X
```

```python
y = read_csv['diagnosis']
```

```python
y
```

# splitting the dataset usung sklearn

```python
from sklearn.model_selection import train_test_split
```

```python

X_train,X_test,y_train,y_test= train_test_split(X,y,test_size = 0.2,random_state=5)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train,y_train)
```

# Evaluating the model

```python
y_predict = svc_model.predict(X_test)
```

```python
y_predict
```

```python
a = confusion_matrix = (y_predict,y_test)
```

```python
sns.heatmap(a,annot = True)
```

# Improving Table

```python
min_train = X_train.min()
```

```python
range_train = (X_train-min_train).max()
```

```python
X_train_scaled = (X_train-min_train)/range_train
```

```python
sns.scatterplot(x = X_train['area_mean'], y = X_train['smoothness_mean'],hue =y_train)
```

```python
sns.scatterplot(x = X_train_scaled['area_mean'], y = X_train_scaled['smoothness_mean'],hue =y_train)
```

```python
min_test = X_test.min()
range_test = (X_test-min_test).max()
X_test_scaled = (X_test-min_test)/range_test
```

```python
svc_model.fit(X_train_scaled,y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
```

```python
a = confusion_matrix = (y_predict,y_test)
```

```python
sns.heatmap(a,annot=True)
```

```python
print(classification_report(y_test,y_predict))
```
